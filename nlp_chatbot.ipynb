```python
import os
from langchain import HuggingFaceHub, PromptTemplate, LLMChain
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

# Set up Hugging Face API token
os.environ['HUGGINGFACEHUB_API_TOKEN'] = 'your_hf_token_here'

def ai_assistant():
    model_id = 'HuggingFaceH4/zephyr-7b-beta'
    conv_model = HuggingFaceHub(
        huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'],
        repo_id=model_id,
        model_kwargs={'temperature': 0.2, 'max_new_tokens': 400}
    )

    template = '''You are an artificial intelligence assistant, answer the question. {question}\n'''
    prompt = PromptTemplate(template=template, input_variables=['question'])

    conv_chain = LLMChain(llm=conv_model, prompt=prompt, verbose=False)

    q = input("Hi, I'm Dusty ğŸ¤–! Ask me a question here:\n")
    response = conv_chain.run(q)

    # Extract and print only the answer
    answer = response.split("\n")[2:]
    print(*answer, sep='\n')

def blog_generator():
    model_id = 'HuggingFaceH4/zephyr-7b-beta'
    conv_model = HuggingFaceHub(
        huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'],
        repo_id=model_id,
        model_kwargs={'temperature': 0.2, 'max_new_tokens': 800}
    )

    template = '''You are a Blog Generator, generate an impressive blog based upon the question. {question}'''
    prompt = PromptTemplate(template=template, input_variables=['question'])

    conv_chain = LLMChain(llm=conv_model, prompt=prompt, verbose=False)

    q = input("Hi, I'm Dusty ğŸ¤–! Ask me a question here:\n")
    response = conv_chain.run(q)

    # Extract and print only the answer
    answer = response.split(". ")[2:]
    print(*answer, sep='\n')

def caption_generator():
    model_id = 'HuggingFaceH4/zephyr-7b-beta'
    conv_model = HuggingFaceHub(
        huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'],
        repo_id=model_id,
        model_kwargs={'temperature': 0.2, 'max_new_tokens': 400}
    )

    template = '''You are a Caption Generator, generate a catchy caption based upon the question. {question}'''
    prompt = PromptTemplate(template=template, input_variables=['question'])

    conv_chain = LLMChain(llm=conv_model, prompt=prompt, verbose=False)

    q = input("Hi, I'm Dusty ğŸ¤–! Ask me a question here:\n")
    response = conv_chain.run(q)

    # Extract and print only the answer
    answer = response.split(". ")[2:]
    print(*answer, sep='\n')

def financial_plan_manager():
    model_id = 'HuggingFaceH4/zephyr-7b-beta'
    conv_model = HuggingFaceHub(
        huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'],
        repo_id=model_id,
        model_kwargs={'temperature': 0.2, 'max_new_tokens': 500}
    )

    template = '''You are a financial planning for young professionals covering topics such as budgeting, saving strategies, investment options, and retirement planning, generate a financial planning based upon the question. {question}'''
    prompt = PromptTemplate(template=template, input_variables=['question'])

    conv_chain = LLMChain(llm=conv_model, prompt=prompt, verbose=False)

    q = input("Hi, I'm Dusty ğŸ¤–! Ask me a question here:\n")
    response = conv_chain.run(q)

    # Extract and print only the answer
    answer = response.split(". ")[2:]
    print(*answer, sep='\n')

def question_generator():
    model_id = 'HuggingFaceH4/zephyr-7b-beta'
    conv_model = HuggingFaceHub(
        huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'],
        repo_id=model_id,
        model_kwargs={'temperature': 0.2, 'max_new_tokens': 600}
    )

    template = '''You are a Question Generator, generate questions based upon the question. {question}'''
    prompt = PromptTemplate(template=template, input_variables=['question'])

    conv_chain = LLMChain(llm=conv_model, prompt=prompt, verbose=False)

    q = input("Hi, I'm Dusty ğŸ¤–! Ask me a question here:\n")
    response = conv_chain.run(q)

    # Extract and print only the answer
    answer = response.split(". ")[2:]
    print(*answer, sep='\n')

def main_menu():
    while True:
        print("\n\nMenu:")
        print("\n\n1) AI Assistant")
        print("2) Blog Generator")
        print("3) Caption Generator")
        print("4) Financial Plan Manager")
        print("5) Question Generator")
        print("6) Exit")

        choice = input("\n\nEnter your choice: ")

        if choice == '1':
            ai_assistant()
        elif choice == '2':
            blog_generator()
        elif choice == '3':
            caption_generator()
        elif choice == '4':
            financial_plan_manager()
        elif choice == '5':
            question_generator()
        elif choice == '6':
            print("\n\nExiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main_menu()
```
